```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
x=np.linspace(-5,5,100)
y=np.sin(x)

sns.set()
for i in range(26):
    a=str(i+1) + '  ' + plt.style.available[i]
    print (a)
```

```python
plt.style.use('seaborn-whitegrid')
plt.plot(x,y)
```
