# Representation graphique du dataset hubble avec Numpy et Matplotlib

```python
# parametres
plt.rcParams.update({ 'font.size' : 15})
```

```python

# line boir et pointiller, label du graph
x=hubble.distance
y=hubble.recession_velocity
print('x min :',np.min(x),'x max :',np.max(x),'y min :',np.min(y),'y max:',np.max(y))
```

```python

plt.plot(x,y, color='black',linestyle='dotted', label='hubble data')
# numerotation des axes Xmin, Xmax, Ymin,Ymax
plt.axis([0,2.2,-450,1100])
#titre
plt.title('Hubble data')
#legende
plt.legend(loc='lower left')
#axes
ax = plt.axes()
ax.set(xlabel='distance', ylabel='r√©cession velocity')
```

# Model

```python
model =  linear_model.LinearRegression()
```

```python
model.fit(x_train,y_train)
```
